### What test is appropriate for this problem? Does CLT apply?

A one-sample test is appropriate for this problem.  We'll be comparing the rate of call backs (# of calls/number of applicants) when the name appears to be black-sounding against the rate of call backs white-sounding names received.

The CLT does apply because the sample size is large enough (>30 samples) as seen above.

### What are the null and alternate hypotheses?

The null hypothesis is that there is no difference in call back rate between black and white sounding names.  The alternative hypothesis is that there is a difference in the call back rate between black and white sounding names.

### Compute margin of error, confidence interval, and p-value.

```python
b_rate = sum(data[data.race=='b'].call)/len(data[data.race=='b'].call)
w_rate = sum(data[data.race=='w'].call)/len(data[data.race=='w'].call)
emp_diff_rate = w_rate - b_rate
print(b_rate, w_rate)
print(emp_diff_rate)
```

We can see here that there IS a difference in the rate which black sounding names are called back in comparison to white sounding names.  Since we're testing the hypothesis that both black-sounding and white-sounding names have the same call back rate, we need to give them the same mean.

```python
mean_call_rate = sum(data.call)/len(data.call)
b_rate_shift = data[data.race=='b'].call - b_rate + w_rate
print(np.mean(b_rate_shift))
```

Next we need to draw bootstrap replicates for both sets of shifted data.  We'll define two functions: diff_from_mean and draw_bs_reps.  Diff_from_mean will find the difference between the mean of an array and the w_rate and draw_bs_reps will use our shifted data and the diff_from_mean function to create an array of many new sample rates assuming we repeated the experiment many, many times.  In effect this lets us recreate the experiment thousands of times using a limited amount of data!

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = func(np.random.choice(data, size=len(data)))

    return bs_replicates
```

```python
def diff_from_mean(data, set_rate=0.0965092402464):
    """Takes the mean of an array and finds the difference against established data"""
    
    return np.mean(data) - set_rate
```

```python
diff_obs = diff_from_mean(b_rate)
bs_replicates = draw_bs_reps(b_rate_shift, diff_from_mean, 10000)

print(diff_obs)
```

With the data above we can now look for statistical significance. We'll assume the usual confidence interval of 95% and look at margin of error, then determine a p-value.

```python
confidence = np.percentile(bs_replicates, [2.5, 97.5])

p = np.sum(bs_replicates <= diff_obs) / len(bs_replicates)

print('At 95% confidence the margin of error is: ', confidence)
print('The p-value is: ', p)
```

### Write a story describing the statistical significance in the context or the original problem.

All of this makes sense within the context of our original observation of the data.  The original mean difference between the two samples was 0.03, which is well outside our 95% confidence interval.  This implies that it is highly UNLIKELY that we would see this number by chance.  In fact when simulating the experiment 10,000 times, we didn't see this result again even once!  This implies that it's very likely black-sounding names get fewer calls for jobs than white-sounding names.
